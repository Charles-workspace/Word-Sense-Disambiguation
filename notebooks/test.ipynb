```python
import pickle
from pathlib import Path


DATA_DIR = Path("data/preprocessed")


with open(DATA_DIR / "corpus.pkl", "rb") as f:
    corpus = pickle.load(f)

with open(DATA_DIR / "token_index.pkl", "rb") as f:
    token_index = pickle.load(f)


words = ["chair", "cloud", "music", "forest"]

print("--- TOKEN FREQUENCIES ---")
for w in words:
    count = token_index.get(w, 0)   # safer than token_index[w]
    print(f"{w:10s}: {count}")
```

```python
def show_context(corpus, doc, sent, tok):
    sent_tokens = corpus[doc][sent]
    target = sent_tokens[tok]
    L1 = sent_tokens[tok-1] if tok-1 >= 0 else None
    R1 = sent_tokens[tok+1] if tok+1 < len(sent_tokens) else None
    L2 = sent_tokens[tok-2] if tok-2 >= 0 else None
    R2 = sent_tokens[tok+2] if tok+2 < len(sent_tokens) else None
    print(f"... {L2} {L1} [{target}] {R1} {R2} ...")

for (doc, sent, tok) in token_index["car"][:10]:
    show_context(corpus, doc, sent, tok)
```
